```python
df_classes["class_labeled"] = df_classes["class"].replace({"2": "licit", "1": "illicit"}) 

class_counts = df_classes["class_labeled"].value_counts()
normalized_class_counts = df_classes["class_labeled"].value_counts(normalize=True)

class_analysis_df = pd.DataFrame({
    "Raw Class Names": df_classes["class"].unique(),
    "Labeled Class Names": df_classes["class_labeled"].unique(),
    "Total Counts": class_counts.values,
    "Total Percentage": normalized_class_counts.values 
})
class_analysis_df
```
